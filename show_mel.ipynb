```python
import os
import matplotlib.pyplot as plt
import numpy as np

mel_folder_path = "C:/Users/admin/Desktop/Narrify/sv2tts_korean/synthesizer/saved_models/balanced_samples_tranfer_learning_best/mel-spectrograms"
gt_mel_folder_path = "C:/Users/admin/Desktop/Narrify_data/balanced_SV2TTS/synthesizer/mels"

def show_mel(mel_path):
    plt.figure(figsize=(4, 16))
    mel = np.load(mel_path)
    plt.imshow(mel)
    plt.show()
    plt.close()

def show_gt_and_generated_mel(gt_mel_path, generated_mel_path):
    gt = np.load(gt_mel_path)
    generated = np.load(generated_mel_path)
    
    # 두 이미지 플롯
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # 첫 번째 이미지 (원본 크기 유지)
    axes[0].imshow(gt, aspect='auto')
    axes[0].set_title(f'Grount truth: {gt.shape}')
    axes[0].set_xlim([0, gt.shape[1]])  # x축을 이미지의 가로 크기에 맞춤
    axes[0].set_ylim([gt.shape[0], 0])   # y축을 이미지의 세로 크기에 맞춤 (위->아래로 출력)

    # 두 번째 이미지 (원본 크기 유지)
    axes[1].imshow(generated, aspect='auto')
    axes[1].set_title(f'Generated: {generated.shape}')
    axes[1].set_xlim([0, generated.shape[1]])  # x축을 이미지의 가로 크기에 맞춤
    axes[1].set_ylim([generated.shape[0], 0])   # y축을 이미지의 세로 크기에 맞춤 (위->아래로 출력)

    # 레이아웃 조정 (각 이미지 비율 유지, 여백 최소화)
    plt.subplots_adjust(wspace=0.5)  # 이미지 간의 간격 조정
    plt.show()
```

```python
gt_mel_path = gt_mel_folder_path+"/mel-KsponSpeech_546311.npy"
mel_path = mel_folder_path+"/mel-prediction-step-5207_sample_1.npy"
show_gt_and_generated_mel(gt_mel_path, mel_path)
```
